使用opencv读取图片


cv2.imread(arg1, arg2)

    arg1:mandatory, path to the image file

    arg2:optional, flags to read the image with a default value of cv2.IMREAD_COLOR or 1

        1：cv2.IMREAD_COLOR:default, reads the image in BGR color format    
        0：cv2.IMREAD_GRAYSCALE:reads the image in grayscale format        
        -1：cv2.IMREAD_UNCHANGED:reads the image as is, including the alpha channel if present   

```python
#read as grayscale
cb_img = cv.imread('checkerboard_18x18.png', cv.IMREAD_GRAYSCALE)

# Print the image data (pixel values), element of a 2D numpy array.
# Each pixel value is 8-bits [0,255]
print(cb_img)
```

attributes of the image

    shape:returns the shape of the image as a tuple (height, width, channels) for color images, or (height, width) for grayscale images.

    dtype:returns the data type of the image, such as uint8 for 8-bit images.

        size:returns the total number of pixels in the image, which is equal to height * width * channels for color images, or height * width for grayscale images.

        

```python
print("Image shape is:",cb_img.shape)
print("Image dtype is:",cb_img.dtype)
print("Image size is:",cb_img.size)
```

Display Images using Matplotlib

```python
plt.imshow(cb_img)
```

What happened?

    Even though the image was read in as a gray scale image, it won't necessarily display in gray scale when using imshow(). matplotlib uses different color maps and it's possible that the gray scale color map is not set.

```python
#set the color map to gray scale for proper rendering
plt.imshow(cb_img, cmap='gray')
```

```python
reversed_cola_img = cola_img[:,:,::-1]
plt.imshow(reversed_cola_img)
```

Splitting and Merging Color Channels

    cv2.split()

        Divides a multi-channel array into several single-channel arrays.

    cv2.merge()

        Merges several arrays into a single multi-channel array.

        !!!All the input matrices must have the same size.

```python
#split the image into the blue, green, and red channels
img_NZ_bgr = cv.imread('New_Zealand_Lake.jpg')
b, g, r = cv.split(img_NZ_bgr)

#show channels
plt.figure(figsize=(20, 5))
plt.subplot(1, 5, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.subplot(1, 5, 2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.subplot(1, 5, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')

plt.subplot(1, 5, 4)
plt.imshow(img_NZ_bgr[:,:,::-1])
plt.title('Original Image')

Merged_img = cv.merge([b, g, r])#merge the channels into a BGR image
plt.subplot(1, 5, 5)
plt.imshow(Merged_img[:,:,::-1])#show the merged image in RGB format
plt.title('Merged Image')

```

Converting to different Color Spaces

    cv2.cvtColor(src,code)

        src:input image: 8-bit unsigned, 16-bit unsigned ( CV_16UC... ), or single-precision floating-point.
        eg: cv2.imread(,cv2.IMREAD_COLOR)
links:https://docs.opencv.org/3.4/d8/d01/group__imgproc__color__conversions.html#ga397ae87e1288a81d2363b61574eb8cab

        code:flag to convert the image to a different color space
links:https://docs.opencv.org/4.5.1/d8/d01/group__imgproc__color__conversions.html#ga4e0972be5de079fed4e3a10e24ef5ef0

            cv2.COLOR_BGR2GRAY:converts the image from BGR to grayscale
            cv2.COLOR_BGR2HSV:converts the image from BGR to HSV
            cv2.COLOR_BGR2RGB:converts the image from BGR to RGB

```python
img_NZ_rgb = cv.cvtColor(img_NZ_bgr, cv.COLOR_BGR2RGB)
plt.imshow(img_NZ_rgb)

```

```python
img_hsv = cv.cvtColor(img_NZ_bgr, cv.COLOR_BGR2HSV)
# split the image into its HSV channels
h,s,v = cv.split(img_hsv)

#show the HSV channels
plt.figure(figsize=(12, 3))
plt.subplot(141)
plt.imshow(h, cmap='gray')
plt.title('Hue channel')
plt.subplot(142)
plt.imshow(s, cmap='gray')
plt.title('Saturation channel')
plt.subplot(143)
plt.imshow(v, cmap='gray')
plt.title('Value channel')

plt.subplot(144)
plt.imshow(img_hsv, cmap='gray')
plt.title('HSV image')

```

Modifying individual Channel

```python
h_new = h + 10
img_hsv_new = cv.merge([h_new, s, v])
img_rgb_new = cv.cvtColor(img_hsv_new, cv.COLOR_HSV2RGB)

plt.figure(figsize=(12, 3))
plt.subplot(141)
plt.imshow(h_new, cmap='gray')
plt.title('hue new_channel')
plt.subplot(142)
plt.imshow(s, cmap='gray')
plt.title('Saturation channel')
plt.subplot(143)
plt.imshow(v, cmap='gray')
plt.title('Value channel')

plt.subplot(144)
plt.imshow(img_rgb_new)
plt.title('RGB new_image')
plt.show()
```
