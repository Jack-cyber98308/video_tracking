```python
cola_img = cv2.imread("coca-cola-logo.png", 1)
plt.imshow(cola_img[:, :, ::-1])
plt.show()
```

Drawing a Line
links:https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2
    cv2.line(img, pt1, pt2, color, thickness, lineType, shift)

        img:input image

        pt1:coordinates of the first point (x1, y1)
        pt2:coordinates of the second point (x2, y2)

        color:line color in BGR format (e.g., (0, 255, 0) for green)

        thickness:line thickness in pixels

        lineType:type of the line (e.g., cv2.LINE_8 for 8-connected lines)
        
        shift:number of fractional bits in the coordinates

```python
cola_line = cola_img.copy()
print(cola_line.shape)
#the line of color is red(recall that opencv uses BGR color format)
img = cv2.line(cola_line, (0, 0), (500, 500), (255, 0,255), 8)
plt.imshow(img[:, :, ::-1])
plt.show()
```

Drawing a circle
links:https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670

    cv2.circle(img, center, radius, color, thickness, lineType, shift)

```python
circle_img = cola_img.copy()
print(circle_img.shape)
#the center of the circle is (400, 400)
#the radius of the circle is 100 半径大小
#the color of the circle is yellow(recall that opencv uses BGR color format)
#the lineType is cv2.LINE_AA(通常是 Anti-Aliased line抗锯齿线)
img = cv2.circle(circle_img, (400, 400), 100, (0, 255,255), -8,lineType=cv2.LINE_AA)
plt.imshow(img[:, :, ::-1])
plt.show()
```

Drawing a rectangle
    
    cv2.rectangle(img, pt1, pt2, color, thickness, lineType, shift)

        img:input image
        pt1:coordinates of the top-left corner of the rectangle
        pt2:coordinates of the bottom-right corner of the rectangle
        color:color of the rectangle in BGR format
        thickness:thickness of the rectangle outline in pixels
        lineType:type of the line used to draw the rectangle
        shift:number of fractional bits in the coordinates of the rectangle vertices

```python
rect_img = cola_img.copy()
print(rect_img.shape)
#the top-left corner of the rectangle is (80, 240)
#the bottom-right corner of the rectangle is (630, 420)
#the color of the rectangle is green(recall that opencv uses BGR color format)
#the lineType is cv2.LINE_AA(通常是 Anti-Aliased line抗锯齿线)
img = cv2.rectangle(rect_img, (80, 240), (630, 420), (0, 255,0), 8,lineType=cv2.LINE_AA)
plt.imshow(img[:, :, ::-1])
plt.show()
```

Add a text

    cv2.putText(img, text, org, fontFace, fontScale, color, thickness, lineType, bottomLeftOrigin)

        img:input image

        text:the text to be written on the image
        
        org:coordinates of the bottom-left corner of the text string in the image
        
        fontFace:font type, such as cv2.FONT_HERSHEY_SIMPLEX, cv2.FONT_HERSHEY_PLAIN, etc.
        
        fontScale:font scale factor that is multiplied by the font-specific base size
        
        color:color of the text in BGR format

        optional parameters:
        
        thickness:thickness of the text outline in pixels
        lineType:type of the line used to draw the text
        bottomLeftOrigin:specifies whether the origin of the text string is the bottom-left corner (True) or the top-left corner (False)


```python
text_img = cola_img.copy()
text = "OpenCV"
org = (300, 400)
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 2
color = (0, 255, 0)
thickness = 2
lineType = cv2.LINE_AA
bottomLeftOrigin = False

cv2.putText(text_img, text, org, fontFace, fontScale, color, thickness, lineType, bottomLeftOrigin)
plt.imshow(text_img[:, :, ::-1])
plt.show()
```
